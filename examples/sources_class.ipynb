```python




# Access the first source in the catalog
first_source = catalog[0]

# Print basic information about the first source
print(first_source)

# Print detailed information about the source including position and spectrum
print(first_source.info(info="all"))

# Get the sky model associated with the source
sky_model = first_source.sky_model()
print(f"Sky model: {sky_model}")

# Access the flux points of the source and print them
flux_points = first_source.flux_points
print(flux_points)

# Display the differential flux points table
flux_table = first_source.flux_points_table
print(flux_table)

# Fit a spectral model to the source
spectral_model = first_source.spectral_model()
if spectral_model:
    print(f"Fitted Spectral Model: {spectral_model}")
else:
    print("Spectral model fitting was not successful or unsupported model type.")
```

```python
from feupy.catalog import SourceCatalogVTSCat

# Initialize the VTSCat catalog using the provided file path
catalog = SourceCatalogVTSCat()

# Access the first source in the catalog
second_source = catalog[0]

# Print basic information about the first source
print(second_source)

# Print detailed information about the source including position and spectrum
print(second_source.info(info="all"))

# Get the sky model associated with the source
obs_info = second_source.observation_info
print(f"Observation info: {obs_info}")

# Access the flux points of the source and print them
flux_points = second_source.flux_points
print(flux_points)

# Display the differential flux points table
flux_tables = second_source.flux_points_tables
print(flux_tables)

# Fit a spectral model to the source
second_source.reference_id_info
```
