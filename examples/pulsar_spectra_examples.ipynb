#### pulsar_spectra: A pulsar flux density catalogue and spectrum fitting repository (https://all-pulsar-spectra.readthedocs.io/en/latest/)

### Example Usage

#### 1. Retrieve Flux Points Table

```python
from feupy.catalog.pulsar_spectra import get_flux_points_table

# Get the flux points table for a specific pulsar
table = get_flux_points_table("J1023+0038")
if table:
    print("Flux Points Table:")
    print(table)
```

#### 2. Retrieve Grouped Tables

```python
from feupy.catalog.pulsar_spectra import get_tables

# Get grouped tables for the pulsar
grouped_tables = get_tables("J1023+0038")
for group in grouped_tables:
    print("Grouped Table Metadata:")
    print(group.meta)
    print(group)
```

#### 3. Retrieve Datasets

```python
from feupy.catalog.pulsar_spectra import get_datasets

# Get datasets for the pulsar
datasets = get_datasets("J1023+0038")
print("Datasets Names:")
datasets.names
```
